```python
import  numpy  as np
import pandas  as pd
import matplotlib.pyplot  as plt
```

```python
x=np.linspace(0,1)
x
```

```python
y=np.sin(4*np.pi*x)*np.exp(-3*x)
y
```

```python
t=pd.DataFrame(y,index=x)  #DataFrame
t.head()
```

```python
t.plot()
plt.show()
```
