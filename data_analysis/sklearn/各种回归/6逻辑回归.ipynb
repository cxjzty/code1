```python
import numpy as np
from  sklearn.linear_model import  LogisticRegression #逻辑回归
from  sklearn.datasets  import make_blobs #创造散点数据
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data,target=make_blobs(n_samples=150, #150个点
                       n_features=2, #分开
                       centers=[[0,0],[3,-1],[4,2]], #3个中心
                       cluster_std=0.8) #制造数据
```

```python
data
```

```python
target #分类
```

```python
plt.scatter(data[:,0],data[:,1],c=target) #数据生成
```

```python
lr=LogisticRegression() #创建逻辑回归
lr.fit(data,target)#训练数据
```

```python
x_min,x_max=data[:,0].min()-1,data[:,0].max()+1 #边界
y_min,y_max=data[:,1].min()-1,data[:,1].max()+1  #-1，1容纳边界
```

```python
x,y=np.arange(x_min,x_max,0.01),np.arange(y_min,y_max,0.01)
#切割，1-10，1.01 ，1.02 ，1.03  2.01 2.02
```

```python
xx,yy=np.meshgrid(x,y)#生成表格
```

```python
a=[1,2,3,4]
b=[5,6,7,8]
np.c_[a,b] #旋转链接
```

```python
xx[:10]
```

```python
xy_test=np.c_[xx.ravel(),yy.ravel()] #生成测试数据 ，链接
```

```python
y_new=lr.predict(xy_test) #预测
z=y_new.reshape(yy.shape) #调整形状
```

```python
plt.scatter(data[:,0],data[:,1],c=target,cmap="gray")
```

```python
plt.pcolormesh(xx,yy,z,cmap="rainbow")
```

```python
plt.pcolormesh(xx,yy,z,cmap="rainbow")
plt.scatter(data[:,0],data[:,1],c=target,cmap="gray")
```
