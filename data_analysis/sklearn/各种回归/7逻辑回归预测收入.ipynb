```python
def  edu2float(edu): #任何学历，返回一个数字代表
    return  np.argwhere(X["education"].unique()==edu)[0,0]
```

```python
def  occupation2float(occupation): #任何学历，返回一个数字代表
    return  np.argwhere(X["occupation"].unique()==occupation)[0,0]
```

```python
X["occupation"]=X["occupation"].map(occupation2float)
X["education"]=X["education"].map(edu2float)
```

```python
X_train=X.iloc[:-1000] #训练数据
X_test=X.iloc[-1000:]#测试数据
Y_train=Y.iloc[:-1000] #训练数据
Y_test=Y.iloc[-1000:]#测试数据
```

```python
logic=LogisticRegression()
logic.fit(X_train,Y_train)
```

```python
Ynew=logic.predict(X_test) #用测试数据产生预测结果
```

```python
result=(Y_test==Ynew) #统计相等个数
accuracy=result.sum()/1000 #统计成功率
accuracy
```
