```python
path=r"C:\Users\Tsinghua-yincheng\Desktop\SZday79\code\codedata\return300.csv"
Return300=pd.read_csv(path) #读取数据
Return300
```

```python
from scipy  import stats #导入科学计算模块
```

```python
density=stats.kde.gaussian_kde(Return300.iloc[:,1]) #取出所有行的第二列，列编号为1  gaussian_kde计算概率密度
bins=np.arange(-5,5,0.02)#切割区间
```

```python
bins
```

```python
density
```

```python
Return300["return"].plot()
```

```python
import  matplotlib
#指定默认字体  
matplotlib.rcParams['font.sans-serif'] = ['SimHei']   
matplotlib.rcParams['font.family']='sans-serif'  
#解决负号'-'显示为方块的问题  
matplotlib.rcParams['axes.unicode_minus'] = False   


plt.subplot(211)   
plt.plot(bins,density(bins))  #按照刻度与分布概率绘图
plt.title("return 300csv  概率分布")
```

```python
plt.subplot(211)   
plt.plot(bins,density(bins).cumsum())  #按照刻度与分布概率绘图
plt.title("return 300csv 累积分布")
```
